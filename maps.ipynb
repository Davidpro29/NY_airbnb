```python
import pandas as pd
import numpy as np
import io
import folium

from google.colab import files

import plotly.express as px 
```

```python
file_uploaded = files.upload()
```

```python
df = pd.read_csv( io.BytesIO( file_uploaded['AB_NYC_2019.csv'] ) )
```

```python
df1 = df.dropna()


df1 = df1.loc[df1['availability_365'] > 0, :]
```

```python
df1.head()
```

```python
# Qual é o aluguel mais caro de cara região da cidade de NY? Apenas imóveis para alugar
# Bronx é U$ 800
# Brooklyn é U$ 8000
# Manhattan é U$ 1000
# Queens é U$ 2600
# Staten Island é U$ 625

cols = ['neighbourhood_group', 'price']

map_barras = df1.loc[:, cols].groupby('neighbourhood_group').max().reset_index()

px.bar(
    map_barras, x='neighbourhood_group', y='price'
)

```

```python
# Onde estão localizados os imóveis mais caros da Região? dados com latitude e longitude, comando para desenhar o mapa, colocar os pontos dentro do mapa


cols = ['price', 'neighbourhood_group', 'latitude', 'longitude']

map_plot = df1.loc[:, cols].groupby('neighbourhood_group').max().reset_index()

# Utilizando o folium
map = folium.Map()

for index, location_info in map_plot.iterrows():
  folium.Marker([location_info['latitude'],location_info['longitude'] ], popup=location_info['neighbourhood_group']).add_to(map)


map
```

```python
# Onde está localizado os imóveis pelo seu tipo? conjunto de dados, deesenhar no mapa e colocar os pontos no mapa.

cols = ['neighbourhood_group', 'room_type', 'latitude', 'longitude']

map_plot = df1.loc[:, cols].sample(100)
map_plot.loc[:, 'color'] = 'NA'

map_plot.loc[map_plot['room_type'] == 'Private room', 'color'] = 'darkgreen'
map_plot.loc[map_plot['room_type'] == 'Entire home/apt', 'color'] = 'darkred'
map_plot.loc[map_plot['room_type'] == 'Shared room', 'color'] = 'purple'


map = folium.Map()

for index, location_info in map_plot.iterrows():
  folium.Marker([location_info['latitude'],location_info['longitude'] ], 
                popup=location_info['neighbourhood_group'],
                icon=folium.Icon(color=location_info['color'])).add_to(map)


map
```
